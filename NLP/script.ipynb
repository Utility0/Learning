```python
doc= nlp("Le prix de la baguette est de 1 euro cinquante")
print('Index:\t\t',[token.i for token in doc])
print('Text:\t\t',[token.text for token in doc])
print('is_Alpha:\t',[token.is_alpha for token in doc]) # Est un character Alphanum√©rique
print('is_Punct:\t',[token.is_punct for token in doc]) # Est une poctuation
print('like_Num:\t',[token.like_num for token in doc]) # Est un nombre
```

```python
import spacy
from spacy.matcher import Matcher

nlp = spacy.load('en_core_web_sm')
matcher = Matcher(nlp.vocab)
pattern  = [[{'ORTH':'iPhone'},{'ORTH':'X'}],[{'IS_DIGIT':True},{'lower':'years'},{'lower':'ago'}],[{'LEMMA':'love','POS':'VERB'},{'POS':'DET','OP':'?'},{'POS':'NOUN'}]]
matcher.add('IPHONE_PATTERN',pattern)

doc = nlp("New iPhone X release date leaked 12 years ago. I loved this smartphone and I love cars")

matches = matcher(doc)

for match,start,end in matches:
    span = doc[start:end]
    print(span.text)
```
